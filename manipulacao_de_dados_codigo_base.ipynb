#### **Selecionando registros de postos do Rio de Janeiro com Preços acima de 2 reais**

````python
data_copy.head(3)
````

````python
data_copy['ESTADO'] == 'RIO DE JANEIRO'
````

````python
data_copy['NÚMERO DE POSTOS PESQUISADOS'] > 100
````

````python
# Juntando as duas condições com o conectivo &
selecao_2 = (data_copy['ESTADO'] == 'RIO DE JANEIRO') & (data_copy['NÚMERO DE POSTOS PESQUISADOS'] > 100)
selecao_2
````

Note que o resultado da seleção continua sendo uma _Series de booleans_ com o _mesmo número de linhas/amostras do DataFrame_, de modo que cada linha possuirá um valor booleano indicando se o posto é do Rio de Janeiro **E** o preço aferido do combustível é maior que 2 reais (True) ou não.

O símbolo **&** significa **AND** na comparação. Essa nomenclatura do python/pandas é diferente das nomenclaturas tradicionais (&&). <br/>
Similarmente:
- **|** representa o **OR** (não é ||)
- **~** representa o **NOT** (não é !)

````python
data_copy[selecao_2].reset_index(drop=True)
````

Alternativamente, poderíamos usar o método `query` para fazermos tal seleção. Porém, isso não é possível especificamente para esse caso, pois o rótulo da coluna 'PREÇO MÉDIO REVENDA' possui caracteres inválidos para o método (cedilha, acentos) 

````python
# Não funciona
# data.query('ESTADO=="RIO DE JANEIRO" and PREÇO MÉDIO REVENDA > 2')
````

````python
data_copy.query('ESTADO =="RIO DE JANEIRO" or ESTADO == "SAO PAULO"').head(3)
````

**Aprofundando mais ainda**

A primeira comparação `(data['ESTADO'] == 'RIO DE JANEIRO')` checa, linha a linha (amostra a amostra) do DataFrame, quais são aquelas cujo o estado é RIO DE JANEIRO. Nenhuma averiguação de preços é feita nesse momento. Como resultado, temos uma Series de booleans que responde **apenas** a essa "pergunta" feita.

A segunda comparação `(data['PREÇO MÉDIO REVENDA'] > 2)` checa, linha a linha (amostra a amostra) do DataFrame, quais são os registro cujo preço do combustível é maior que 2 reais. Note que essa comparação checará os postos de **TODOS os estados**. Como resultado, temos uma Series de booleans que responde **apenas** a essa "pergunta" feita.

Por fim, as duas "perguntas" são unidas pelo AND (&) que retorna a "pergunta completa" que fizemos.

Alguns podem argumentar que tal abordagem é **ineficiente**, uma vez que, para cada condição ("pergunta"), estamos varrendo todas as linhas do DataFrame. <br/>
O Pandas _tenta otimizar_ isso ao máximo por de trás dos panos. Mas, de fato, de tivermos um dataset **muito grande** (centenas de milhares de linhas), tal abordagem se tornará _lenta_.

Assim, poderíamos fazer filtragem com múltiplos condicionais em partes:

````python
selecao_3 = data_copy['ESTADO'] == 'PERNAMBUCO'
postos_pe = data_copy[selecao_3]
postos_pe.head(3)
````

````python
selecao_4 = data_copy['NÚMERO DE POSTOS PESQUISADOS'] > 600
selecao_4.shape
````

````python
postos_pe_segunda_Variavel = postos_pe[selecao_3].reset_index(drop=True)
postos_pe_segunda_Variavel
````

#### **Selecionando registros de postos de São Paulo ou do Rio de Janeiro com Gasolina Comum acima de 2 reais**

Podemos fazer a solução do "jeito mais lento", percorrendo o DataFrame inteiro _múltiplas vezes_:

````python
data_copy.columns
````

````python
sp_ou_rio = (data_copy['ESTADO'] == 'SAO PAULO') | (data_copy['ESTADO'] == 'RIO DE JANEIRO')
produto_gaso_comun = (data_copy['PRODUTO'] == 'GASOLINA COMUM')
preco_medio_revenda = (data_copy['PREÇO MÉDIO REVENDA'] > 2)
final = sp_ou_rio & produto_gaso_comun & preco_medio_revenda
chama = data_copy[final].reset_index(drop=True)
````

````python
print(chama['ESTADO'].unique())
print(chama['PRODUTO'].unique())
print(chama['PREÇO MÉDIO REVENDA'].min())
````

<br/>

````python
postos_sp_rj = data_copy[sp_ou_rio]
postos_sp_rj_gasolina = postos_sp_rj[produto_gaso_comun]
postos_sp_rj_gasolina_preco_medio = postos_sp_rj_gasolina[preco_medio_revenda]
````

````python
print(postos_sp_rj.shape)
print(postos_sp_rj_gasolina.shape)
print(postos_sp_rj_gasolina_preco_medio.shape)
````

**Criando uma função 'def' para imprimir esse mapinha de corr quando o pai quiser // economizar tempo e código**

````python
def mapa_corr(dataframe):
    # Conhecendo as info Primeiro
    print(dataframe.info())
    
    # Criando a correlação:
    dataframe_2 = dataframe.corr(numeric_only=True)

    # Criando o heatmap usando o Seaborn:
    plt.figure(figsize=(10, 8))
    sns.heatmap(dataframe_2, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Heatmap de Correlação')
    plt.show()

    ## Teste para ajudar a observação: ##

    # Criar uma máscara booleana para filtrar os valores entre -1 e -0.5, ou entre 0.5 e 1
    mascara = ((dataframe_2 >= -1) & (dataframe_2 <= -0.5)) | ((dataframe_2 >= 0.5) & (dataframe_2 <= 1))

    # Aplicar a máscara convertendo a matriz de correlação em uma série usando stack()
    correlacao_filtrada = dataframe_2[mascara].fillna('<>')

    # Verificar se existem valores filtrados
    if not correlacao_filtrada.empty:
        # Use describe() para obter estatísticas descritivas dos valores filtrados
        descricao_correlacao_filtrada = correlacao_filtrada.describe()

        # Visualize os valores filtrados
        display(correlacao_filtrada)
        print(correlacao_filtrada.shape)
        display(descricao_correlacao_filtrada)
        print(descricao_correlacao_filtrada.shape)
    else:
        print("Nenhum valor de correlação encontrado na faixa desejada.")
````

````python
# Função gráfico linha

def graf_linha(df, coluna1, coluna2):
    x = df[coluna1]
    y = df[coluna2]

    # Plotando o gráfico linear com grid
    plt.plot(x, y, marker='o', linestyle='-', color='b')
    plt.xlabel(coluna1)
    plt.ylabel(coluna2)
    plt.title(f'{coluna1} X {coluna2}')
    plt.grid(True)  
    plt.show()
````

````python

def plot_graf(df, coluna1, coluna2, tipo='linear'):
    x = df[coluna1]
    y = df[coluna2]

    if tipo == 'linear':
        # Plotando o gráfico linear com grid
        plt.plot(x, y, marker='o', linestyle='-', color='b')
        plt.xlabel(coluna1)
        plt.ylabel(coluna2)
        plt.title('Gráfico Linear')
        plt.grid(True)
    elif tipo == 'scatter':
        # Plotando o gráfico de dispersão (scatter plot) com grid
        plt.scatter(x, y, marker='o', color='b')
        plt.xlabel(coluna1)
        plt.ylabel(coluna2)
        plt.title('Gráfico de Dispersão')
        plt.grid(True)
    elif tipo == 'histograma':
        # Plotando o histograma com grid
        plt.hist(x, bins=10, edgecolor='black')
        plt.xlabel(coluna1)
        plt.ylabel('Frequência')
        plt.title('Histograma')
        plt.grid(True)
    else:
        print("Tipo de gráfico inválido. Escolha entre 'linear', 'scatter' ou 'histograma'.")

    plt.show()

# Exemplos de uso:
# Supondo que você tenha um DataFrame chamado 'df' com as colunas 'coluna1' e 'coluna2'
# plot_graph(df, 'coluna1', 'coluna2', tipo='linear')
# plot_graph(df, 'coluna1', 'coluna2', tipo='scatter')
# plot_graph(df, 'coluna1', 'coluna2', tipo='histograma')
````
